```python
from cv2 import imread, imshow, rectangle, cvtColor, COLOR_BGR2RGB
from cv2 import CascadeClassifier
import matplotlib.pyplot as plt
```

```python
def showFaces(image):
    #Loading Image
    pixels = imread(image)

    #Loading pretrained model from OpenCV Github Library
    classifier = CascadeClassifier("../input/haarcascade-frontalface-defaultxml/haarcascade_frontalface_default.xml")    

    #Detecting Images
    boundary_points = classifier.detectMultiScale(pixels,1.05,8)

    #Tuned the model to scaleFactor 1.05 and minNeighbours 8

    for point in boundary_points:
        #Getting detected co-ordinates
        x, y, width, height = point
        x2, y2 = x + width, y + height
        #Draw a rectangle over the pixels
        rectangle(pixels, (x, y), (x2, y2), (0,0,255), 1)

    #Showing the detected faces
    img = cvtColor(pixels,COLOR_BGR2RGB)
    plt.imshow(img)
    plt.show()
```

```python
showFaces('../input/faceimages/test1.jpg')
showFaces('../input/faceimages/test2.jpg')
```
